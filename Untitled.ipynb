```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set
```

```python
train_data = pd.read_csv('train (1).csv')
test_data = pd.read_csv('test (1).csv')
test_result = pd.read_csv('sample_submission.csv')
train_data.head()
```

```python
print(train_data.shape)
print(test_data.shape)
print(test_result.shape)
```

```python
test_data['SalePrice'] = test_result['SalePrice']
test_data.head()
```

```python
all_data = pd.concat([train_data, test_data], axis = 0, ignore_index= True)
```

```python
test_data.describe()
```

```python
all_data.describe()
```

```python
all_data.isnull().sum().sort_values(ascending=False)
```

```python
all_data = all_data.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence','FireplaceQu'], axis=1)
```

```python
all_data.isnull().sum().sort_values(ascending=False)
```

```python
all_data = all_data.dropna(axis=0)
```

```python
all_data.describe()
```

```python
sns.distplot(all_data['SalePrice'])
```

```python
q = all_data['SalePrice'].quantile(0.99)
```

```python
all_data_1 =all_data[all_data['SalePrice']<q]
all_data_1.describe(include='all')
```

```python
sns.distplot(all_data_1['SalePrice'])
```

```python
all_data_1=pd.get_dummies(all_data_1, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = all_data_1.drop(['SalePrice'], axis = 1)
y = all_data_1['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=365,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
inputs = scaler.fit(X_train)
inputs_scaled = scaler.transform(X_train)
```

```python
reg = LinearRegression()
reg.fit(inputs_scaled, y_train)
```

```python
y_hat = reg.predict(X_test)
```

```python
plt.scatter(y_test, y_hat)
plt.xlabel('Targets (y_test)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
reg.score(X_test, y_test)
```
